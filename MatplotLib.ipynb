```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig_size = [16,9]
```

<b>Çizgi Grafiği(Plot)</b><br/>
2 değişken arasındaki ilişkiyi görselleştirmek için kullanılırlar.<br/>
X ekseni bağımsız değişkeni(girdiyi), Y ekseni ise bağımlı değişkeni(çıktıyı) belirtecek şekilde kullanmak mantıklı olacaktır. Çünkü noktaları birleştirerek bir çizgiler oluşturuyoruz.

```python
x = np.linspace(0,2*np.pi,50)
```

```python
y = np.sin(x)
```

```python
plt.figure()
plt.plot(x,y,label="Sinus") #Fonksiyona(çizgiye) bir label atadık. plt.legend() ile bu label'ı gösteriyoruz
plt.xlabel("X Degerleri") #x eksenine etiket tanımladık
plt.ylabel("Y Degerleri") #y eksenine etiket tanımladık
plt.title("Sinus Grafiği") #grafiğe bir başlık verdik
plt.legend() #plot fonksiyonu içinde label parametresi ile verdiğimiz labelların bilgi olarak gözükmesini sağladık
plt.grid() #grafiği soluk çizgilerle parçalara böldük
plt.show() #jupyter notebook dışındaki ortamların çoğunda plt.show() yazmazsak grafik gözükmez
```

<b>Plot Fonksiyonu Parametreleri</b><br/>
<li><b>color:</b>Çizginin rengini belirtir. Green, Blue, b, g, #81179b gibi değerler alır</li>
<li><b>linestyle:</b>Çizginin şeklini belirtir. '-', '--', '-.', ':', '' gibi değerler alır</li>
<li><b>linewidth:</b>Çizginin kalınlığını veren bir reel sayı değeri alır.</li>
<li><b>marker:</b>Herbir x noktası ve onun y değeri için grafikte karşılık gelen noktayı işaretler. '.', 'o' gibi bir çok değer alır</li>
<li><b>markerfacecolor:</b>Markerın rengini belirtir.</li>
<li><b>markersize:</b>Markerın büyüklüğünü belirtir.</li>
<li><b>markeredgewidth:</b>Markerın kenarlarının genişliğini belirtir</li>
<li><b>markeredgecolor:</b>Markerın kenarlarının rengini belirtir</li>

```python
plt.figure(figsize=fig_size)

#plt.plot(x,y,color="green",marker="o",linestyle="--")
#Aşağıdaki kod ile yukarıdaki kod aynı işi yapıyor.
#x ve y paramlarından sonra veriler string türünden paramyukarıdaki kodu kısa yoldan yapmamızı sağladı
plt.plot(x,y,"go--") # "Renk_kısaltmasıMarkerÇizgi_tipi" -- yerleri değişebilir, bazıları olmayabilir
```

<b>Pyplot Fonksiyonlari(plt.fonk)</b><br/>
<li><b>xticks/yticks:</b>plot fonksiyonuna verdiğimiz x ve y değerleri ile grafiğimizi çiziyorduk. Default olarak bir ayarlama yapmazsak grafikde de bu çizdirmek için kullandığımız değerler gözükecektir. xticks/yticks fonksiyonları ile bu değerlerin yerine kendi belirlediğimiz değerlerin gelmesini sağlarız. "ticks" parametresi etiketlenmek istenen değerleri belirten bir listedir. "labels" parametresi ise bu değerlere karşılık gelen etiket listesidir. ticks boş olabilir ama labels boş olamaz. labels boş bırakılırsa grafikte sadece tickse yazılan değerlere gözükür. labels parametresi de olursa sadece bu değerlere karşılık gelen labeller gözükür.</li>
<li><b>annotate:</b>Grafik üzerindeki bir noktayı işaretlememizi sağlar.</li>

```python
plt.figure(figsize=fig_size)

#market='o',color="blue",linestyle="-"
plt.plot(x,y,'ob-')

#kesikli çizgi ile kırmızı renkte, çizgi kalınlığı 5
plt.plot(x,y*0.3,'--r',linewidth=5)

plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.title("Sinüs Grafiği")

#eğer plot fonksiyonunda label değerlerini vermemişsek bu şekilde de legend bilgilerini yazdırabiliriz.
#Verilen listenin sırası plot fonksiyonlarının çağrılma sırası ile uyumlu olmalıdır.
plt.legend(["Normal Sinüs","0.3 ile Çarpılmış Sinüs"])

#buradaki labellerin yerine direkt olarak pi işareti yazmak için kullanılan en iyi yol latex dilidir.
#Bu dili bir ara öğrenmek lazım
plt.xticks(ticks=[0,np.pi/2,np.pi,1.5*np.pi,2*np.pi],labels=["0","pi/2","pi","1.5pi","2pi"])

#mesaj,okunbaşlangıçnoktaları,okunbitişnoktaları,okungörselparametreleri
#bu fonksiyon kullanılacağı zaman dökümanı inceleyerek yapmak daha mantıklı akılda tutulmasına gerek yok.
#mesaj kısmında pi sayısını yazarken latex dilinden faydalandık
plt.annotate("Tam bu noktaya bakın!\n ($\pi$,0)",[np.pi,0],[1.5*np.pi,0.6],
            arrowprops=dict(facecolor="cyan",shrink=0.05,edgecolor="green"))

plt.grid()
plt.show()
```

<b>Eşik Değeri Grafiği</b>

```python
y = np.array([2,3,4,5,7,9,12,13,14,10,7,11,14,15])
x = np.arange(len(y))
plt.plot(x,y,color="black")
esik = 9
altEsik = y < esik
plt.scatter(x[altEsik],y[altEsik],color="black")
ustEsik = np.logical_not(altEsik)
plt.scatter(x[ustEsik],y[ustEsik],color="red")
plt.axhline(esik,color="red",linestyle="--")
plt.show()
```

<b>Nokta Grafiği(Scatter)</b><br/>
Genelde 2 bağımsız değişken arasındaki ilişkiyi göstermek için kullanılır. Örnek olarak x ve y ekseni 2 farklı girdi niteliğini ifade ederken, noktaların rengini farklı tutarak da sınıfları belirtebiliriz.<br/>
<b>Scatter plotta iki değişkenin ilişkileri gösterilirken(genelde x ve ye eksenine bağımsız değişkenleri koyup, işaretlemenin rengini,şekline göre bağımlı değişkenin özelliklerini belirtiriz.); gözlem noktaları bir bulut şeklinde, yapısal olmayan bir şekilde dağılmışsa bu iki değişken arasında ilişki yoktur veya çok zayıftır diyebiliriz. Eğer gözlem noktaları belli bölgelerde gruplanmış ise bağımlı değişkenin sınıflarına göre grafik daha da detaylandırılıp(farklı sınıflara farklı renkler/sınıflar atama gibi) yorumlanmalıdır.</b>

<b>x ve y ekseninde değişkenlerimiz olsun. Bu doğruları fit eden doğrunun eğimi sıfır veya sıfıra çok yakınsa bu iki değişken arasındaki ilişki(korelasyon) yoktur veya çok zayıftır.</b>

Kendi renklerimizi vermek istersek aşağıdaki gibi yapabiliriz:

```python
"""
Alternatif olarak verisetimizi her bir sınıf için birer parçaya bölerek de aşağıdaki grafiği çizdirebilirdik.
"""

colors = ['r','g','b']

plt.figure(figsize=fig_size)

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

for i,name in enumerate(iris.target_names):
    plt.scatter(x=iris.data[iris.target == i,0],y=iris.data[iris.target==i,1],s=iris.data[iris.target==i,2]*20,c=colors[i])
plt.legend(iris.target_names)
plt.show()
```

Buradaki sorunumuz legend kısmındaki işaretlerin büyüklüklerinin farklı olması. Şimdi onu düzeltmeye çalışacağız:

```python
colors = ['r','g','b']

plt.figure(figsize=fig_size)

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

#noktaların büyüklüğünü belirtmek için boş scatter çizdiriyoruz. Sadece legend kısmını ayarlamak için
for area in [1.5, 4, 6]:
    plt.scatter([], [], c='k', alpha=0.3, s=area*20,
                label=str(area) + ' cm')

#yukarıdaki cell ile aynı burası
for i,name in enumerate(iris.target_names):
    plt.scatter(x=iris.data[iris.target == i,0],y=iris.data[iris.target==i,1],s=iris.data[iris.target==i,2]*20,c=colors[i],
               alpha=0.4,label=iris.target_names[i]) #alpha parametresi ile saydamlığı belirliyoruz. üst üste inen noktalar daha koyu olacak.
                          #bu sayede bu noktaları tespit edebiliriz.
legend = plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='Petal Length(cm)/Tür')      

#sınıfları belirten legendlerin boyutlarını 30'a eşitledik
legend.legendHandles[-1]._sizes = [30]
legend.legendHandles[-2]._sizes = [30]
legend.legendHandles[-3]._sizes = [30]

plt.show()
```

<b>3 Boyutlu Sütun Grafik</b>

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

emisyon = [['60070','55832','51178'],
          ['66790','61821','70085'],
          ['120668','113903','132248']]
emisyon = np.array(emisyon,dtype=np.int)
sekil = plt.figure()
ax = Axes3D(sekil)

xlabels = np.array(['2012','2013','2014'])
xpos = np.arange(xlabels.shape[0])
ylabels = np.array(['Konut','Üretim','Enerji'])
ypos = np.arange(ylabels.shape[0])
a,b = np.meshgrid(xpos,ypos,copy=False)
zpos = emisyon
zpos = zpos.ravel()

dx=0.5;dy=0.5;dz=zpos

ax.w_xaxis.set_ticks(xpos)
ax.w_xaxis.set_ticklabels(xlabels)
ax.w_yaxis.set_ticks(ypos+0.6)
ax.w_yaxis.set_ticklabels(ylabels)

values = np.linspace(0.2,1.,a.ravel().shape[0])
renkler = cm.rainbow(values)
ax.bar3d(a.ravel(),b.ravel(),dz*0,dx,dy,dz,color=renkler)
plt.title("Türkiye'nin üç yıllık seragazı emisyon değerleri",fontsize=9,fontname="Cambria")
plt.show()
```

<b>Grup Sütun Grafik</b><br/>
bar(np.arange(3),[10,20,30]) fonksiyonunu ele alalım. x parametresine 0'dan 3'e kadar tamsayı dizisi verdik. Bu sayılara denk eksen değerlerinin ortasında sütunlar çizilecektir. Sütunun genişliğinin yarısı o noktanın sağında diğer yarısı da solunda olacaktır. Bu bilgiler ışığında aşağıdaki gibi bir grup sütun grafiği çizebiliriz.<br/>
Aşağıdaki örnek ilk grubun A sütunu için sütunun orta noktasının değeri -0.2, B için 0.2 olacaktır. Genişliği de 0.3 olarak seçtik.<br/>
A ürünü için bunun 0.15'lik kısmı -0.2 ile -0.35(-0.2-0.15) arasında diğer kısmı da -0.2 ile -0.05(-0.2+0.15) arasında olacaktır.<br/>
B ürünü içinde orta noktamız 0 + 0.2 den 0.2 olacaktır. Bu grafikde de aynı şekilde 0.15 kısım solda 0.15 kısımda bu orta noktanın sağında yer alacaktır.

```python
plt.figure(figsize=fig_size)
y_1 = [10,5,3,12,20]
y_2 = [5,8,9,10,11]
x_labels = ["Ocak","Şubat","Mart","Nisan","Mayıs"]
x_pos = np.arange(len(x_labels))

plt.bar(x=x_pos-0.2,height=y_1,color="r",width=0.3)
plt.bar(x=x_pos+0.2,height=y_2,width=0.3)

plt.xticks(ticks=x_pos,labels=x_labels)
plt.xlabel("Aylar")
plt.ylabel("Satış Miktarları (x1000)")
plt.legend(["A Ürünü","B Ürünü"])

plt.show()
```

<b>Çekirdek Yoğunluk Tahmini Grafiği(Kernel Density Estimate, KDE)</b><br/>
Histogram benzeri bir işlevi vardır. Histogramdan farklı olarak sürekli bir şekilde veri dağılımını gösterirler.<br/>
Farklı fonksiyonlar ile yoğunluk tahmin grafikleri çizilebilir. Bu örnekte Gauss fonksiyonu kullanıldı.

```python
from scipy import stats
```

```python
x = np.random.randint(20,80,50)

gkde = stats.gaussian_kde(x) #gauss fonksiyonu olarak çizdireceğiz
i = np.linspace(10,100,100)
kdepdf = gkde.evaluate(i)
plt.figure()
plt.hist(x,bins=15,normed=1,color="#bbadf9")
plt.plot(i,kdepdf,label="Çekirdek Yoğunluk",color="b")
plt.grid(True)
plt.legend()
plt.show()
```

### Kontür Grafik

```python
delta = 0.025
x = np.arange(-3,3,delta)
y = np.arange(-2,2,delta)
X,Y = np.meshgrid(x,y)

Z = X**2 + Y**2

fig,ax = plt.subplots()
CS = ax.contour(X,Y,Z,levels=np.arange(0,14,1))
plt.clabel(CS,inline=1,fontsize=10)
plt.show()
```

```python
fig1,ax1 = plt.subplots()
out = ax1.pcolor(X,Y,Z)

col_bar = fig1.add_axes([0.2,0.1,0.5,0.05])
fig1.colorbar(out,cax=col_bar)

plt.show()
```
